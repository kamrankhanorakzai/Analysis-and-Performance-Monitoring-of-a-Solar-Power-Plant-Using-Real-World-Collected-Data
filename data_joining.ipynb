# Merge Data 1

```python

df_unit=None
for i in range(1,13):
     if i>=9 and i<=10:
            continue
     year=2025       
     if i>8:
        year=2024
        temp_df=pd.read_excel("{0}-{1}.xlsx".format(year,i),sheet_name=0,names=['date',"E-Today(KWH)"],usecols=[0,1],skiprows=3)
        df_unit=pd.concat([df_unit,temp_df],ignore_index=True)
        continue
     temp_df=pd.read_excel("{0}-0{1}.xlsx".format(year,i),sheet_name=0,names=['date',"E-Today(KWH)"],usecols=[0,1],skiprows=3)
     df_unit=pd.concat([df_unit,temp_df],ignore_index=True)
 
```

```python
df_unit["date"]=pd.to_datetime(df_unit["date"])
df_unit 
```

```python
df_unit.info()
```

Feature extraction

```python
df_unit["months"]=df_unit.date.dt.month_name()
df_unit["days"]=df_unit.date.dt.day_name()
df_unit["years"]=df_unit.date.dt.year
df_unit["week"]=df_unit.date.dt.isocalendar().week
df_unit["DayOfWeek"]=df_unit.date.dt.weekday
df_unit['IsWeekend'] = df_unit['date'].dt.weekday >= 5
df_unit['Quarter'] = df_unit['date'].dt.quarter
df_unit["date"]=pd.to_datetime(df_unit["date"]).dt.date

df_unit
```

# Merge Data 2

```python
df_error=pd.DataFrame()
for i in range(0,44):
     temp_df=pd.read_excel("error_2025-09-16 ({0}).xlsx".format(i),usecols=["time","event"])
     df_error=pd.concat([df_error,temp_df],ignore_index=True)

df_error.rename(columns={"time":"time/date"},inplace=True)    
df_error
```

```python
df_error.info()
```

Feature extraction

```python
df_error["time/date"]=pd.to_datetime(df_error["time/date"])
df_error["date"]=df_error["time/date"].dt.date
df_error["time"]=df_error["time/date"].dt.strftime('%H:%M:%S')

df_error["months"]=df_error["time/date"].dt.month_name()
df_error["days"]=df_error["time/date"].dt.day_name()
df_error["years"]=df_error["time/date"].dt.year
df_error["week"]=df_error["time/date"].dt.isocalendar().week
df_error["DayOfWeek"]=df_error["time/date"].dt.weekday
df_error['IsWeekend'] = df_error["time/date"].dt.weekday >= 5
df_error['Quarter'] = df_error["time/date"].dt.quarter

df_error
```
