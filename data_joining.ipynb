# Merge Data 2

```python
df_error=pd.DataFrame()
for i in range(0,44):
     temp_df=pd.read_excel("data/error_2025-09-16 ({0}).xlsx".format(i),usecols=["time","event"])
     df_error=pd.concat([df_error,temp_df],ignore_index=True)

df_error.rename(columns={"time":"time/date"},inplace=True)
df_error["time/date"]=pd.to_datetime(df_error["time/date"])
df_error["date"]=pd.to_datetime(df_error["time/date"].dt.date)

df_error
```
